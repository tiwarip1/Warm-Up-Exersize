```python
import math
import matplotlib.pyplot as plt
%matplotlib inline

der_l=[]

def derivative_2c(h,x):
    der=(function(x+h)-function(x))/h
    return der
    
def function(x):
    return math.atan(x)

for h in range(1,1000):
    der_l.append(derivative_2c(1/h,math.sqrt(2)))

plt.plot(der_l)
```

```python
err_der=[]

def err_2c(value,expected):
    for item in value:
        err_der.append(math.log10((expected-item)/expected))

err_2c(der_l,(1/3))
plt.plot(err_der)
```

```python
der_3c=[]

def derivative_3c(h,x):
    return math.log10(function(x+h)-function(x-h))/(2*h)

for h in range(1,1000):
    der_3c.append(derivative_3c(h,math.sqrt(2)))
    
plt.plot(der_3c)
```

```python
err_der3c=[]

def error_3c(expected,value):
    for item in value:
        err_der3c.append(((expected-item)/expected))
        
error_3c((1/3),der_3c)
plt.plot(err_der3c)
```
